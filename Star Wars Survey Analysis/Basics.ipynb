# Overview

This project would try to answer the question 
"Does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?"

This question was posed by a team at FiveThirtyEight when they asked 835 Star Wars fasn this question via SurveyMonkey. This question was asked prior to the release of the film 'Star Wars: The Force Awakens' 

This data set would need to be cleaned before it can be used as i have found a lot of empty slots 

```python
#Reading in the dataset

import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')

star_wars.head()
```

```python
#Removing rows without RespondentID

star_wars = star_wars[pd.isnull(star_wars['RespondentID'])== False]

star_wars.head()
```

### Cleaning and Mapping Yes/No Columns

```python
#Converting some replies from 'YES' and 'NO' answers to 'True' or 'False' 
#making use of thepd.series.map() method
import numpy as np
yes_no ={'Yes': True,
          'No': False,
          np.NaN: False }

#'Have you seen any of the 6 films in the Star Wars franchise?' column
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] =star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
#'Do you consider yourself to be a fan of the Star Wars film franchise?' column
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] =star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

star_wars.head()
```

### Cleaning and Mapping Checkbox Columns

```python
#Changing the answers for 'Which of the following Star Wars films have you seen? Please select all that apply.' to 'True' or 'False'
   
#star_wars = star_wars.columns[3:9]
import numpy as np

#creating map
true_false = { 'Star Wars: Episode I  The Phantom Menace': True,
               'Star Wars: Episode II  Attack of the Clones': True,
               'Star Wars: Episode III  Revenge of the Sith':True,
               'Star Wars: Episode IV  A New Hope':True,
               'Star Wars: Episode V The Empire Strikes Back':True,
               'Star Wars: Episode VI Return of the Jedi':True,
               np.NaN: False   
             }

columns = star_wars.columns[3:9]
for elem in columns:
    star_wars[elem] = star_wars[elem].map(true_false)

star_wars.head()
```

```python
#Renaming the columns

star_wars = star_wars.rename(columns = {'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
                  'Unnamed: 4': 'seen_2',
                  'Unnamed: 5':'seen_3',
                  'Unnamed: 6':'seen_4',
                  'Unnamed: 7':'seen_5',
                  'Unnamed: 8':'seen_6'
})
star_wars.head()
```

```python
#Converting and renaming the columns[9:15].

#converts to float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns= {
                'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
                'Unnamed: 10':'ranking_2',
                'Unnamed: 11':'ranking_3',
                'Unnamed: 12':'ranking_4',
                'Unnamed: 13':'ranking_5',
                'Unnamed: 14':'ranking_6'
})

#star_wars.columns[9:15]
star_wars.head()
```

N.B:

- ranking_1: Star Wars: Episode I The Phantom Menace

- ranking_2: Star Wars: Episode II Attack of the Clones

- ranking_3: Star Wars: Episode III Revenge of the Sith

- ranking_4: Star Wars: Episode IV A New Hope

- ranking_5: Star Wars: Episode V The Empire Strikes Back

- ranking_6: Star Wars: Episode VI Return of the Jedi

### Finding the highest-ranked movie

```python
import matplotlib.pyplot as plt 
%matplotlib inline

columns =['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5', 'ranking_6']

#Plot a bar chart of the rankings of the movies
(star_wars[columns].mean()).plot.bar() 

plt.title('Movie Ranking')
plt.xlabel("Movie #")
plt.ylabel('Average Ranking')
```

### Finding the most watched movie in the franchise

```python
 movie = ['seen_1', 'seen_2', 'seen_3','seen_4', 'seen_5', 'seen_6']

#Bar Plot
(star_wars[movie].sum()).plot.bar()

plt.title('Film Most Viewed')
plt.xlabel('Movie #')
plt.xlabel('Number That Viewed it')
```

### Findings So Far

On average it looks like the respondents chose ranking_5 ('Star Wars: Episode V The Empire Strikes Back') has the best star wars movie so far. 

Star Wars: Episode III Revenge of the Sith (ranking_3) was voted overwhelmingly as the worst movie in the Star wars franchise


Most Viewed:

It looks like the fifth installment('Star Wars: Episode V The Empire Strikes Back') in the franchise is the most viewed movie.

Overall the last two movies has seen a spike in interest in the franchise has more people have seen it than anyother ones.

The third installment though been the highest rated is surprisingly the least viewed movie.
