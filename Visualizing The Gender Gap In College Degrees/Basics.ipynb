# Overview

The main focus of this project is plotting asthetics. We are simply going to look at the number of women in various majors between 1968-2011. We will then plot our findings to see if there are majors with less women than others.

It should be noted that this project has nothing to do with the gender wage gap.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

#Making the code firndly to colur blind people using Tableaus specification
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

#creating a list of an ordered list(ordered in descedning order form largest gap to smallest gap) 
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    
    #removes the border lines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    
    #names the titles
    ax.set_title(stem_cats[sp])
    
    #removes excess lines at the border
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    #This is annotaions for naming the lines
    #placed at the leftmost and rightmost graph
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
plt.show()
```


Here the columns(Majors) are grouped into STEM, Liberal Arts and others. Here i tried to reduce the clutter on the plot to the barest minimum.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(20,24))

#STEM Majors
for elem in range(0,6):
    ax2 = fig.add_subplot(6,3,3*elem+1)
    ax2.plot(women_degrees['Year'], women_degrees[stem_cats[elem]], c= cb_dark_blue, label = 'Women', linewidth = 3)
    ax2.plot(women_degrees['Year'], 100-women_degrees[stem_cats[elem]], c= cb_orange, label = 'Men', linewidth = 3)
        
    ax2.set_title(stem_cats[elem])
    ax2.set_xlim(1970,2010)
    ax2.set_ylim(0,100)
        
    ax2.tick_params(top ='off', bottom = 'off', right= 'off', left ='off', labelbottom ='off')
            
    #removing spines from the chart for cleare
    for key,spine in ax2.spines.items():
        spine.set_visible(False)

    #set annotations(labelling the lines) on the chart
    if elem == 0:
        ax2.text(2005,15,'Men')
        ax2.text(2005,85,'Women')
    if elem == 5:
        ax2.text(2005,89,'Men')
        ax2.text(2005,8,'Women')
        ax2.tick_params(labelbottom='on')
    #this makes the graph only display 100 and 0 on the y axis
    ax2.set_yticks([0,100])
    
    #to make it easier to see we put in an horizontal line across and made it slightly transparent
    ax2.axhline(50,c=(171/255,171/255, 171/255), alpha= 0.5) 
        
#liberal Major
for elem in range(0,5):  
        ax2 = fig.add_subplot(6,3,3*elem+2)
        #STEM Majors
        index = 0
        ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[elem]], c= cb_dark_blue, label = 'Women', linewidth = 3)
        ax2.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[elem]], c= cb_orange, label = 'Men', linewidth = 3)
        
        ax2.set_title(lib_arts_cats[elem])
        ax2.set_xlim(1970,2010)
        ax2.set_ylim(0,100)
        
        ax2.tick_params(top ='off', bottom = 'off', right= 'off', left ='off', labelbottom = 'off')
        
        #removing spines from the chart for cleare
        for key,spine in ax2.spines.items():
            spine.set_visible(False)
        
        #set annotations(labelling the lines) on the chart
        if elem == 0:
            ax2.text(2005,20,'Men')
            ax2.text(2005,75,'Women')
        if elem == 4:
            ax2.tick_params(labelbottom='on')
            
        #this makes the graph only display 100 and 0 on the y axis
        ax2.set_yticks([0,100])
        
        #to make it easier to see we put in an horizontal line across and made it slightly transparent
        ax2.axhline(50,c=(171/255,171/255, 171/255), alpha= 0.5) 

# other majors          
for elem in range(0,6):
    ax2 = fig.add_subplot(6,3,3*elem+3)
    ax2.plot(women_degrees['Year'], women_degrees[other_cats[elem]], c= cb_dark_blue, label = 'Women', linewidth = 3)
    ax2.plot(women_degrees['Year'], 100-women_degrees[other_cats[elem]], c= cb_orange, label = 'Men', linewidth = 3)
        
    ax2.set_title(other_cats[elem])
    ax2.set_xlim(1970,2010)
    ax2.set_ylim([0,100])
        
    ax2.tick_params(top ='off', bottom = 'off', right= 'off', left ='off', labelbottom ='off')
            
    #removing spines from the chart for cleare
    for key,spine in ax2.spines.items():
        spine.set_visible(False)

    #set annotations(labelling the lines) on the chart
    if elem == 0:
        ax2.text(2005,9,'Men')
        ax2.text(2005,90,'Women')
    if elem == 5:
        ax2.text(2005,65,'Men')
        ax2.text(2005,35,'Women')
        ax2.tick_params(labelbottom='on')
    
    #this makes the graph only display 100 and 0 on the y axis
    ax2.set_yticks([0,100])
    
    #to make it easier to see we put in an horizontal line across and made it slightly transparent
    ax2.axhline(50,c=(171/255,171/255, 171/255), alpha= 0.5)            


#This export all of the plots contained as a single image file
plt.savefig('gender_degrees.png')
plt.show()
```
