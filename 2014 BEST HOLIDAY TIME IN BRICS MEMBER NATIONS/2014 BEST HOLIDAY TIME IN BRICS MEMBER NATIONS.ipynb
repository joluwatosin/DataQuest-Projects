## Getting the data

Weather Underground keeps historical weather data collected in many airports around the world.
Weather data was obtained from The Open University's Learn to code for Data Analysis course for learning purpose and has a csv format.


### Data Cleaning  

```python
import pandas as pd

#Reading in file
brasilia = pd.read_csv('Brasilia_BSB_2014.csv', skipinitialspace = True)
moscow = pd.read_csv('Moscow_SVO_2014.csv', skipinitialspace = True)
delhi = pd.read_csv('Delhi_DEL_2014.csv', skipinitialspace = True)
beijing = pd.read_csv('Beijing_PEK_2014.csv', skipinitialspace = True)
cape_town = pd.read_csv('CapeTown_CPT_2014.csv', skipinitialspace = True)

#Editing Columns
brasilia = brasilia.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
brasilia['WindDirDegrees'] = brasilia['WindDirDegrees'].str.rstrip('<br />')
brasilia['WindDirDegrees'] = brasilia['WindDirDegrees'].astype('int64')
brasilia['Date'] = pd.to_datetime(brasilia['Date']) #converting the date column to datetime type

moscow = moscow.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].str.rstrip('<br />')
moscow['WindDirDegrees'] = moscow['WindDirDegrees'].astype('int64')
moscow['Date'] = pd.to_datetime(moscow['Date']) #converting the date column to datetime type

delhi = delhi.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
delhi['WindDirDegrees'] = delhi['WindDirDegrees'].str.rstrip('<br />')
delhi['WindDirDegrees'] = delhi['WindDirDegrees'].astype('int64')
delhi['Date'] = pd.to_datetime(delhi['Date']) #converting the date column to datetime type

beijing = beijing.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
beijing['WindDirDegrees'] = beijing['WindDirDegrees'].str.rstrip('<br />')
beijing['WindDirDegrees'] = beijing['WindDirDegrees'].astype('int64')
beijing['Date'] = pd.to_datetime(beijing['Date']) #converting the date column to datetime type

cape_town = cape_town.rename(columns={'WindDirDegrees<br />':'WindDirDegrees'})
cape_town['WindDirDegrees'] = cape_town['WindDirDegrees'].str.rstrip('<br />')
cape_town['WindDirDegrees'] = cape_town['WindDirDegrees'].astype('int64')
cape_town['Date'] = pd.to_datetime(cape_town['Date']) #converting the date column to datetime type

```

## Brasilia, Brazil


```python
%matplotlib inline
brasilia.index = brasilia['Date'] #changing the index to Date

#Mean Temperature graph; checking for temperatures between 16 - 24 degrees
brasilia['Mean TemperatureC'].plot(grid= True, figsize=(10,5))
```

```python
#Percipitation graph; i.e. Level of rainfall
brasilia['Precipitationmm'].plot(grid= True, figsize=(10,5))
```

### Findings So Far
Based on the graphs above, it would be better if we look for holiday dates between January to end of March. 
The summer months in Brazil are December to end of March
This fulfills the two criteria needed:
- Low rainfall
- Average Temperatures of around 22-24 degrees

```python
summer = brasilia[pd.datetime(2014,1,1):pd.datetime(2014,4,30)]
summer[['Mean TemperatureC','Precipitationmm']].plot(grid=True, figsize=(10,5))
```

```python
#This looks at January 20 to February 20
summer_holiday = brasilia[pd.datetime(2014,1,20):pd.datetime(2014,2,20)]
summer_holiday[['Mean TemperatureC','Precipitationmm']].plot(grid=True, figsize=(10,5))
```

### Brasilia Holiday Dates
The best time for a 2 weeks holiday would be between the 27th of January and the 12 of February.
This days have no rainfall and have good temperatures. Of course this is no guarantee that the weather pattern will repeat itself in future years. To make a sensible prediction we would need to analyse the summers for many more years.
